__What is Apache Spark and how is it different to Apache Hadoop? __

Apache Spark is a cluster computing environment for big data processing which provides an interface for programming entire clusters with implicit data parallelism and fault-tolerance. It’s different from Hadoop in a few ways:
- It leverages in-memory processing instead disk-only data flow. This should drastically speed up processing time as RAM is incredibly fast compared to disk I/O.
- Uses RDDs and dataframes, both make it more convenient to interact with data.
- It utilizes a lazy evaluation scheme, which means it'll only perform processing chains once an action (collect function) is called upon that chain.

# HW10.4.1: Making Homegrown KMeans more efficient

The above provided homegrown KMeans implentation in not the most efficient. How can you make it more efficient? Make this change in the code and show it work and comment on the gains you achieve. 

### HINT: have a look at  [this linear regression notebook](    http://nbviewer.jupyter.org/urls/dl.dropbox.com/s/atzqkc0p1eajuz6/LinearRegression-Notebook-Challenge.ipynb)

````python
import numpy as np

#Calculate which class each data point belongs to
def nearest_centroid(x):    
    closest_centroid_idx = np.sum((x - centroids.value)**2, axis=1).argmin()
    return (closest_centroid_idx,(x,1))

def DistancetoCenter(p):
    return np.sqrt(np.sum((p-centroids.value)**2, axis=1).min())

K = 3
#centroids = np.array([[0.0,0.0],[2.0,2.0],[0.0,7.0]])
centroids = sc.broadcast(np.array([[0.0,0.0],[2.0,2.0],[0.0,7.0]]))

D = sc.textFile("data.csv").map(lambda line: np.array([float(x) for x in line.split(',')])).cache()
for i in range(100):  
    res = D.map(nearest_centroid).reduceByKey(lambda x,y : (x[0]+y[0],x[1]+y[1])).collect()
    res = sorted(res,key = lambda x : x[0])
    
    # average by cluster size
    centroids_new = np.array([x[1][0]/x[1][1] for x in res])  
    centroids = sc.broadcast(centroids_new)
    
    #for simplicity, only print iterations we were examining in previous problems
    if (i+1) in [1, 10, 20, 30, 40, 50, 100]:
        print "\nIteration %d" %(i+1)
        print centroids.value
        plot_iteration(centroids.value)
        WSSSE = D.map(DistancetoCenter).reduce(lambda x, y: x + y)
        print("Within Set Sum of Squared Error = " + str(WSSSE))
print "\nFinal Results:"
print centroids.value
````

I've made it more efficient by borrowing the use of the Broadcast function used in the other notebook. Since I'm using Broadcast, it greatly reduces the communication cost in Spark. We do see a slightly better experience with this improved class (more noticeable in the higher iterations), but the time difference would be even more drastic with bigger datasets and/or a greater number of iterations or clusters.
