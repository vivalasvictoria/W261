## HW11.0: Broadcast versus Caching in Spark 


Q: __What is the difference between broadcasting and caching data in Spark? Give an example (in the context of machine learning) of each mechanism (at a highlevel). Feel free to cut and paste code examples from the lectures to support your answer.__

<u>Broadcasting</u> - When all tasks of a job require access to the same static data (like in a map-side join), Spark will ship the data to each task in the closure. If there are 100 tasks on a 10 node cluster, then the same data is shipped to a node 10 times. With broadcasting, the data is sent to a node only once.  
  
<u>Caching</u> - When you're running an iterative job, the same training data is needed in all the iterations. Without caching, all the data will be read from disk at each iteration. With caching, the entire dataset is read only once.

In the context of logistic regression describe and define three flavors of penalized loss functions.  Are these all supported in Spark MLLib (include online references to support your answers)?

#### Loss Terms
* Zero-one - This is very basic. It only counts the number of mistakes. It doesn't appear to be supported in MLLib. 

$$
L_{01}(m) = \left\{
        \begin{array}{ll}
            0 & \quad m \geq 0 \\
            1 & \quad m < 0
        \end{array}
    \right.
$$

* Hinge loss - This calculates the loss term for soft-margin SVM. It is probably not very important for logistic regression, but it is a supported loss function in MLlib.

$$L_{hinge} = R_2(w) + \sum L_{hinge}(m_i) $$


* Log loss - This is the one we care most about for logistic regression. It's the equivalent to the kind of loss function used to train a logistic regression model. This is supported in MLlib.

$$J(w) = \lambda ||w||^2 + \sum log 1 + e^{-y^{(i) f_w(x^{(i)})}} $$


* Squared loss - This one does exactly what it describes - square the error value. It is supported in MLLib.

$$L_2(m) = (f_w(x) - y)^2 = (m-1)^2$$

Source: 

https://spark.apache.org/docs/2.1.0/mllib-linear-methods.html#loss-functions

Only one of these accurately applies to logistic regression, so let's look at the normalization components that are supported to aid the loss function.


#### MLLib Supported Logistic Loss + Penalty Functions

* Logistic Loss with Elastic Net - 
Elastic net is a function of both L1 and L2. It is not recommended to train models without any regularization, especially when the number of training examples is small.


From the Spark documentation:

$$\alpha||w||_1 + (1-\alpha) \frac{1}{2} ||w||\frac{2}{2}$$

Gradient: $\alpha * sign(w) + w(1-\alpha)$

(Note: sign(w) is the vector consisting of the signs (Â±1) of all the entries of w.)


* Logistic Loss with L1 Normalization (Lasso) - 
L1 regularization can help promote sparsity in weights leading to smaller and more easily interpretable models, which is useful for feature selection. The L1 loss is a function of the absolute values of the weights for the penalty. The tuning parameter, $\lambda$ controls the strength of the penalty term. For $\lambda$ between two extremes, the L1 penalty may cause some coefficients to be shrunken to zero.

$$argmin_w~L(w) = 1/n \sum_{i=0}^{n-1} log(1+e^{-y.w^T.x_i})+ \lambda.|w|$$


From the Spark documentation:

$$||w||_1 $$

Gradient: $sign(w)$

(Note: sign(w) is the vector consisting of the signs (Â±1) of all the entries of w.)


* Logistic Loss with L2 Normalization (Ridge) - 
L2 regularized problems are generally easier to solve than L1 regularized due to the smoothing. Ridge regression shrinks the estimated coefficients. This helps to avoid situations in which a single coefficient completely dominates the outcome of the regression. The tuning parameter, $\lambda$, controls the strength of the penalty term. Bias increases as $\lambda$ increased; while variance decreases as $\lambda$ increases.

$$argmin_w~L(w) = 1/n \sum_{i=0}^{n-1} log(1+e^{-y.w^T.x_i})+ \lambda.w^T.w $$


From the Spark documentation:

$$\frac{1}{2} ||w||\frac{2}{2} $$

Gradient: $w$


References:

https://spark.apache.org/docs/2.1.0/mllib-linear-methods.html#regularizers

There are several differences in L1 and L2 priors for penalized logistic regression:
* L1 tends to drive w to zero more efficiently than L2 normalization
* L2 doesn't tolerate both very large and very small values of coefficients, though L1 does 
* L2 regularization corresponds to Gaussian priors, whereas L1 regularization corresponds to double exponential priors (a Gaussian is the exponential of a quadratic and the double exponential is the exponential of a linear).

````python
import random
from numpy import array
from numpy.random import rand

def generateData(n, percent_nonseparable=-1):
    xb = (rand(n)*2-1)/2-0.5
    yb = (rand(n)*2-1)/2+0.5
    xr = (rand(n)*2-1)/2+0.5
    yr = (rand(n)*2-1)/2-0.5
    inputs = []
    for i in range(len(xb)):
        if(random.random() <= percent_nonseparable):
            inputs.append([xb[i],yb[i],0])
            inputs.append([xr[i],yr[i],1])
        else:
            inputs.append([xb[i],yb[i],1])
            inputs.append([xr[i],yr[i],0])
    return inputs

def plot_data(data, title="A Plot"):
    plt.figure(figsize=(10,10))
    data1 = []
    data2 = []
    for x in data:
        if x[2] == 1:
            data1.append(x)
        elif x[2] == 0:
            data2.append(x)
    xs1 = array([float(x[0]) for x in data1])
    ys1 = array([float(x[1]) for x in data1])
    xs2 = array([float(x[0]) for x in data2])
    ys2 = array([float(x[1]) for x in data2])
    pylab.plot(xs1, ys1, '+', color='blue')
    pylab.plot(xs2, ys2, '*', color='red')
    pylab.title(title)

def plot_lines(allWeights, allIntercept, allLabels):
    for z in xrange(len(allWeights)):
        weights = allWeights[z]
        intercept = 0
        if z < len(allIntercepts):
            intercept = allIntercepts[z]
        label = allLabels[z]
        x = [-1, 1]
        y = [ -(i*weights[0] + intercept)/weights[1] for i in x]
        plt.plot(x, y, label=label)
    plt.legend()
````

````python
def logRegGD(data, test=None, wInitial=None, learningRate=0.05, iterations=50, regParam=0.01, regType="Lasso"):
    w = None
    wReg = None
    lastGradient = None
    featureLen = len(data.take(1)[0].features)
    n = data.count()
    
    if wInitial is None:
        #w could be broadcasted here
        w = np.random.normal(size=featureLen)
    else:
        w = wInitial
        
    weights = []
    intercepts = []
    labels = []
    predictedTotal = 0
    iterationsToPrint = [1, 10, 20, 50, 100, 500, 1000, 1500, 2000]
    
    if test != None:
        predictedTotal = test.count()
        
    for i in range(iterations):
        tolerance = .001
        wBroadcast = sc.broadcast(w)
        gradient = data.map(lambda p: (1 / (1 + np.exp(-p.label*np.dot(wBroadcast.value, p.features)))-1) \
                            * p.label * p.features * np.reciprocal(np.sqrt(np.sum(np.square(p.features))))) \
                       .reduce(lambda x, y: x + y)/n

        if regType == "Ridge":
            wReg = wBroadcast.value * 1
            wReg[-1] = 0
        elif regType == "Lasso":
            wReg = wBroadcast.value * 1
            wReg[-1] = 0 
            wReg = (wReg > 0).astype(int) * 2-1
        else:
            wReg = np.zeros(wBroadcast.value.shape[0])

        wdelta = learningRate * (gradient + regParam * wReg)
        
        #gradient:  GD of error^2 + GD of regularized term 
        gradient = gradient + regParam * wReg
        
        if np.sum(np.abs(wdelta)) <= tolerance*np.sum(np.abs(wBroadcast.value)):
            print "Converged after {} iterations.".format(i+1)
            break
        else:
            delta = np.sum(np.abs(wdelta)) - tolerance*np.sum(np.abs(wBroadcast.value))
        
        w = wBroadcast.value - wdelta
        
        if test != None:
            predictedCorrect = test.map(lambda p: np.dot(w, p.features) - p.label).filter(lambda x: x < 0).count()
            accuracy = float(predictedCorrect) / float(predictedTotal)
            errors = 1 - accuracy
            
            if i+1 in iterationsToPrint:
                print "Iteration ", i+1
                print "------------------"
                print "Weights: ", w
                print "Error: ", errors
                print ""
                weights.append(w)
                labels.append("Iteration {}".format(i+1))
                intercepts.append(0)
                
        lastGradient = gradient
    return (w, (weights, intercepts, labels))
````
