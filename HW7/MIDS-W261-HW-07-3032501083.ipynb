````python
%%writefile MRJob_InitiateAdjacency.py
from mrjob.job import MRJob
from mrjob.step import MRStep
import sys

class initiateAdjacency(MRJob):

    def configure_options(self):
        super(initiateAdjacency, self).configure_options()
        self.add_passthrough_option('--startNode', default='1')

    def mapper(self, _, line):
        #initial mapping of nodes
        cols = line.strip().split('\t')
        node = cols[0]
        neighbors = eval(cols[1])
        if node == self.options.startNode:
            yield node, [neighbors, 0, [node], 'Q']
        else:
            yield node, [neighbors, sys.maxint, [], 'U']

if __name__ == '__main__':
    initiateAdjacency.run()
````

````python
%%writefile MRJob_ShortestPath.py
from mrjob.job import MRJob
from mrjob.step import MRStep
import sys

class shortestPath(MRJob):
    
    def mapper(self, _, line):
        cols = line.strip().split('\t')
        name = str(eval(cols[0]))
        rest = eval(cols[1])
        neighbors = rest[0]
        distance = int(rest[1])
        path = rest[2]
        status = rest[3] 

        # If this node is queued, change to V and continue to next; 
        #else, output
        if status == 'Q':
            yield name, [neighbors, distance, path, 'V']
            if neighbors:
                for node in neighbors:
                    temp = list(path)
                    temp.append(node)
                    yield node, [None, distance + 1, temp, 'Q']
        else:
            yield name, [neighbors, distance, path, status]
 
    def reducer(self, key, values):
        #initialize variables
        path = []
        neighbors = {}
        distance = sys.maxint
        status = None

        for val in values:
            #node has been visited--break the loop.
            if val[3] == 'V':
                path = val[2]
                neighbors = val[0]
                distance = val[1]
                status = val[3]
                break

            #node has not been visited.
            #get neighbors and the status
            #if status != Q, overwrite
            elif val[0]:
                neighbors = val[0]
                if status != 'Q':
                    status = val[3]

            #otherwise, it's unvisisted
            #update status and path
            else:
                path = val[2]
                status = val[3]             

            #update min distance
            distance = min(distance, val[1])

        yield key, [neighbors, distance, path, status]

if __name__ == '__main__':
    shortestPath.run()
````

````python
from MRJob_InitiateAdjacency import initiateAdjacency
from MRJob_ShortestPath import shortestPath

def findShortestPath(filename, startNode, finalNode):
    #temp directory for files
    temp_dir = 'w261_temp/' + os.path.dirname(filename)
    !mkdir -p $temp_dir
    dirFiles = 'w261_temp/' + filename

    #get graph adjacency list
    mr_job_init = initiateAdjacency(args=[
        filename, '-r', 'local',
        '--no-strict-protocols',
        '--startNode', startNode
    ])

    #iterate over adjacency list until all nodes are visited
    with open(dirFiles, 'w') as f:
        with mr_job_init.make_runner() as runner:
            runner.run()

            for line in runner.stream_output():
                out = mr_job_init.parse_output_line(line)
                f.write(str(out[0]) + '\t' + str(out[1]) + '\n')

    filenameTemp = dirFiles + '.tmp'

    #find shortest path
    mr_job = shortestPath(args=[
        dirFiles,
        '--no-strict-protocols'
    ])

    for i in range(10):
        with open(filenameTemp, 'w') as f:
            with mr_job.make_runner() as runner:
                runner.run()

                for line in runner.stream_output():
                    out = mr_job.parse_output_line(line)
                    f.write(str(out[0]) + '\t' + str(out[1]) + '\n')

                    if out[0] == finalNode and out[1][3] == 'V':
                        path = out[1][2]
                        return path

        !mv {filenameTemp} {dirFiles}
````

````python
%%writefile MRJob_EDA.py
from mrjob.job import MRJob
from mrjob.protocol import ReprProtocol
from mrjob.step import MRStep

class EDA(MRJob):

    INPUT_PROTOCOL = ReprProtocol
    INTERNAL_PROTOCOL = ReprProtocol
    OUTPUT_PROTOCOL = ReprProtocol
 
    def configure_options(self):
        super(EDA, self).configure_options()
        self.add_passthrough_option('--EDAtype', default='nodes')
    
    # get count of nodes
    def mapper_discover(self, key, value):
        yield str(key), 1
        for node in value:
            yield str(node), 1

    def reducer_discover(self, key, value):
        yield key, 1

    def mapper_count(self, key, value):
        yield None, 1

    def reducer_count(self, key, value):
        yield None, sum(value)

    
    #get distribution of out degrees
    def mapper_links(self, name, neighbors):
        degree = len(neighbors)
        yield degree, 1

    def reducer_links(self, degree, occurrences):
        yield degree, sum(occurrences)

    def steps(self):
        if self.options.EDAtype == 'nodes':
            return [
                MRStep(mapper = self.mapper_discover,
                       combiner = self.reducer_discover,
                       reducer = self.reducer_discover),
                MRStep(mapper = self.mapper_count,
                       combiner = self.reducer_count,
                       reducer = self.reducer_count)
            ]
        elif self.options.EDAtype == 'links':
            return [
                MRStep(mapper = self.mapper_links,
                       combiner = self.reducer_links,
                       reducer = self.reducer_links)
            ]

if __name__ == '__main__':
    EDA.run()
````

````python
def printPath(indicesfile, path):
    path = path
    path_Dict = {}
 
    with open(indicesfile, 'r') as f:
        for line in f:
            fields = line.strip().split('\t')
            if fields[1] in path:
                path_Dict[fields[1]] = fields[0]

    print '{:<10s}{:<50s}'.format('Index', 'Name')
    print '--------------------'
    distance = 0
    for node in path:
        distance = distance + 1
        print '{:<10s}{:<50s}'.format(node, path_Dict[node])
    print '\nTotal Distance is: ', (distance - 1)
````

````python
from datetime import datetime
from MRJob_InitiateAdjacency import initiateAdjacency
from MRJob_ShortestPath import shortestPath
import subprocess as sp

def findShortestPathHadoop(filename, startNode, endNode):
    global baseDir

    # create an output folder for the files
    outDirName = filename.replace('.txt', '_state2').split('/')[-1]
    outDirState = 'hdfs://%s/%s' % (baseDir, outDirName)
    !hdfs dfs -rm -r -f -skipTrash $outDirState > /dev/null
    
    #print to keep track of where the job is
    print datetime.today(), 'Initializing adjacency list'

    #initiate graph adjacency list
    mr_job_init = initiateAdjacency(args=[
        filename,
        '--strict-protocols',
        '--startNode', startNode,
        '-r', 'hadoop',
        '--cleanup-on-failure=ALL', 
        '--cleanup=ALL',
        '--output-dir', outDirState
    ])

    with mr_job_init.make_runner() as runner:
        runner.run()

    # Iterate over the adjacency list with state until all nodes are visited
    inputDir = outDirState + '/'
    outputDir = outDirState + 'Temp1'

    #initialize variables
    isFinished = False
    path = None
    i = 1

    while not isFinished and path is None:
        !hdfs dfs -rm -r -f -skipTrash $outputDir > /dev/null
        
        #print iteration begin time
        print datetime.today(), 'Beginning iteration', i

        mr_job = shortestPath(args=[
            inputDir,
            '--strict-protocols',
            '-r', 'hadoop',
            '--cleanup-on-failure=ALL', 
            '--cleanup=ALL',
            '--output-dir', outputDir
        ])

        with mr_job.make_runner() as runner:
            runner.run()
            
            #print iteration complete time
            print datetime.today(), 'Completed iteration', i
            isFinished = True
            
            proc = sp.Popen(["hadoop", "fs", "-cat", outputDir + "/*"], stdout=sp.PIPE)
            output = proc.communicate()[0].decode('utf-8')

            # Write stream_output to file
            for line in output.splitlines():
                out = mr_job.parse_output_line(line)
                if out[0] == endNode and out[1][3] != 'U':
                    path = out[1][2]
                elif out[1][3] == 'Q':
                    isFinished = False

        # Update inputDir and outputDir for next iteration
        inputDir = outDirState + 'Temp' + str(i) + '/'
        outputDir = outDirState + 'Temp' + str(i + 1)

        i += 1

    return path
````

````python
%%writefile EDAWikipedia.py
from mrjob.job import MRJob
import json
import sys

class EDAWikipedia(MRJob):
    
    def mapper_init(self, _, line):
        node, neighbors = line.split("\t")
        neighbors = neighbors.replace("'",'"').replace(",",", ")
        neighbors = json.loads(neighbors)
        if node == self.options.source_node:
            yield node, (neighbors, 0, 'Q', '')
        else:
            yield node, (neighbors, sys.maxint, 'U', '')
    
    def configure_options(self):
        super(EDAWikipedia, self).configure_options()
        self.add_passthrough_option(
            '--source_node', default='1')
        self.add_passthrough_option(
            '--directed', default='', choices=['true', 'false', ''])
                       
    def mapper_degree_counter(self, node, values):
        neighbors = values[0]        
        degree_out = 0
        if neighbors:
            degree_out = len(neighbors.keys())
            for n in neighbors.keys():
                yield n, (1, 0)
        yield node, (0, degree_out)
                
    def reducer_degree_counter(self, node, values):
        in_count = 0
        out_count = 0
        for v in values:
            in_count += v[0]
            out_count += v[1]
        yield node, (in_count, out_count) 
    
    def mapper_total_edges(self, node, values):
        yield node, values
        yield "Edge count", (values[0])
        yield "Node count", 1
    
    def reducer_total_edges(self, node, values):
        edgeTotal = 0
        nodeTotal = 0
        
        if node == "Node count":
            for node_count in values:
                nodeTotal +=  node_count
            yield "Total nodes: ", nodeTotal
            
        elif node == "Edge count":
            for edge in values:
                edgeTotal += edge
            if self.options.directed == 'true':
                yield "Total edges: ", edgeTotal
            elif self.options.directed == 'false':
                yield "Total edges: ", edgeTotal/2
            else:
                yield "Total edges [If directed, If undirected]: ", (edgeTotal, edgeTotal/2)
        
        else:
            for d in values:
                yield node, d
    
    def steps(self):
        return [self.mr(mapper = self.mapper_init),

                self.mr(mapper = self.mapper_degree_counter,
                         reducer = self.reducer_degree_counter
                        ),
                self.mr(mapper = self.mapper_total_edges,
                         reducer = self.reducer_total_edges
                        ),
               ] 
    
if __name__ == '__main__':
    EDAWikipedia.run()
````

First, I'd do a reducer side join of the file that contains all the simple paths to the neighbors table, which contains the nodes and all neighbors. This would result in a new table containing the simple paths so that I could compare them. Something like:

Initialization:

- Create a paths table consisting of all the start nodes and simple paths.

Mapper:

- Emit the last node in the simple path as the key and the entire path as the value.
- For the neighbors table, emit the node as the key and the neighbors as the value. Prepend `*` as the first part of the value in order to ensure that the entry appears first in a secondary sort.

Reducer:

- For the join, I'd read in the neighbors first utilizing the secondary sort. If there are no neighbors, we will assume that it is a dangling node and so our list of neighbors is length zero.

    * Otherwise, emit a new entry where the new neighbor is the key and the path resulting from adding the node to the path is the value.

This means we'll end up with a file with all the simple paths that we can compare to find the largest path(s).
